# Visualizing data on how people spend their time during quarantine

## The data:

The data used here is collected by conducting a survey in the Instagram story among my followers. The data collected is then entered in excel sheets and stored as csv file (.csv format)

Now, the data is analysed by visualizing them in bar chart format and pie chart.

### Importing necessary packages:

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Reading the csv file

```python
df_people=pd.read_csv(r'C:\Users\Srividhya\Desktop\quarantinetasks.csv')
print(df_people)
```

#### Specifying only certain columns

```python
df_tasks = pd.DataFrame(df_people, columns= ['Tasks','Count'])
print (df_tasks)
```

## Bar Chart

```python
plt.figure(figsize=(10,6))
# title
plt.title("Average count of the tasks people are doing in this quarantine")
# Bar chart
sns.barplot(x=df_tasks['Tasks'], y=df_tasks['Count'])
plt.xticks(rotation=90)
# Add label for vertical axis
plt.ylabel("Count of tasks performed")
```

## Pie Chart

```python
labels = df_tasks['Tasks']
plt.pie(df_tasks['Count'], labels= labels, autopct='%.1f%%')
plt.show()
```

### Enhancing the pie chart by introducing legends

```python
labels = df_tasks['Tasks']  
fig, ax1 = plt.subplots(figsize = (24,18)) 
ax1.pie(df_tasks['Count'], autopct='%.1f%%', shadow='true') 
plt.title('Average count of the tasks people are doing in this quarantine', fontsize = 24) 
ax1.legend(labels, loc = 'upper left') 
plt.tight_layout() 
plt.show()
```

By visualizing the data,it is clearly seen how different people are spending their time during this lockdown period. Now, it is our choice to change the plot by spending the time correctly.
